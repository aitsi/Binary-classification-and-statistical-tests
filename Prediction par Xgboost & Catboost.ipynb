## Une dépendance statistiquement significative entre :

#### (a) La spécialité et le sexe 

```python
# Table de contingence : tableau de croisement entre les modalités des deux variables
table = pd.crosstab(df.specialite, df.sexe)
table
```

* l’hypothèse nulle est « les deux variables testées sont indépendantes ». Nous allons étudier cette hypothèse à l'aide du test de chi 2

```python
resultat_test = chi2_contingency(table)
print('Statistique de test: ',resultat_test[0])
print('p-value: ',            resultat_test[1])
print('Degré de libeté: ',    resultat_test[2])
```

* l'hypothèse nulle  est rejetée (p<5%) et il y a donc une dépendance statistique entre spécialité et sexe

Mesurons le niveau de dépendance entre les deux variables en utilisant le V de Cramer. Il se calcule à l’aide de la statistique de test, du degré de liberté et des dimensions de la table de contingence. Il renvoie une valeur entre 0 et 1

```python
# table de contingence avec margins
table = pd.crosstab(df.specialite, df.sexe, margins=True)
table
```

```python
from math import sqrt
# Récupération de la taille d’échantillon 
n = 19809 # voir table de contingence avec margins
# Récupération du nombre de modalité de chaque variable
p = 2
q = 4
# Calcul du dénominateur du V de Cramér
m = min(p-1,q-1)
# Calcul du V de Cramér
V = sqrt(resultat_test[0]/(n * m))
print('la valeur de V de Cramer est : ', V)
```

* Comme V=0.37, la dépendance entre spécialité et sexe est plutôt une dépendance significative et modérée

# Part II : Machine Learning

###  Un modèle permettant de prédire la variable embauche, avec optimisation des paramètres

* Le Gradient Boosting est un algorithme d'apprentissage supervisé dont le principe est de combiner les résultats d'un ensemble de modèles plus simples et plus faibles pour fournir une meilleure prédiction. 
* La prédiction de la variable embauche est un problème de classification binaire qui nécessite un algorithme d'apprentissage supervisé, nous allons donc utiliser un classificateur pour prédire cette variable.
* Nous utiliserons les classificateur XGBoost et Catboost qui sont des véritables star des compétitions de Machine Learning et qui sont basés sur des arbres de décision. En effet, la plupart des variables que nous avons sont des variables catégorielles ou des variables entières (même si leur type est float) donc il y aura toujours des profils similaires (nous pouvons toujours trouver des profils avec 2 ans d'expérience et avec d'autres caractéristiques en commun, au contraire si l'expérience est une variable réelle il est difficile de trouver deux candidats avec la même valeur d'expérience). Les modèles basés sur les arbres de décision sont performants sur ce type de données. Le modèle Catboost fait également partie des nouveaux modèles puissants basés sur les arbres de décision.

#### Pré-traitement des données

Nous avons appliqué les deux méthodes les plus connues pour transformer nos variables catégorielles en variables numériques. Il s'avère que le Labelencoding donne de meilleures performances que le OnehotEncoding bien que ce dernier soit plus favorisé puisqu'il n'entraîne pas le modèle à reconnaître un ordre dans les valeurs d'une même variable.

```python
df.columns
```

###### Traitement des variables catégorielles

```python
rate_specialite = (df.groupby('specialite').size()) / len(df)
rate_specialite
```

```python
## frequancy Encoding pour la variable specialite 
df['specialite'] = df['specialite'].apply(lambda x : rate_specialite[x]) 
```

```python
## diplome : Il y a quatre niveaux de dilpome (existance d'un ordre)
## dispo, sexe : juste deux valeurs
cols = ['sexe','dispo','diplome']

# Convertir les colonnes en LabelEncoder
df[cols]= df[cols].apply(LabelEncoder().fit_transform)
```

```python
## One-Hot-Encoding pour la variable specialite (pour ne pas tenir compte de l'ordre)
df = pd.get_dummies(df, 
                    columns=['mois'],
                    drop_first=True)               
```

```python
# Suppression des variables inutiles
y=df.embauche
df=df.drop(['embauche','cheveux','date'], axis=1)
```

###### Train-Test Split

```python
# Définition des données d'entrainement et de test avec un ration de 0.8 
X_train, X_test, y_train, y_test = train_test_split(df, 
                                                    y, 
                                                    random_state=42,
                                                    stratify=y,   ## pour avoir le même pourcentage pour les embauché
                                                                  ## et non embauché
                                                    test_size = 0.2) # train_size = 0.8
X_val, X_validation, y_val, y_validation = train_test_split(df, 
                                                            y, 
                                                            test_size = 0.2, 
                                                            stratify=y, random_state=42)

df.head()
```

Dans un premier temps nous allons appliquer le modèle CatBoost  

#### Application du modèle CatBoost

```python
# catboost classifier
import catboost
from catboost import CatBoostClassifier, cv, Pool

cat_features=['sexe','mois','specialite','diplome','dispo']
cols = ['age', 'exp', 'salaire']
train_pool = catboost.Pool(X_train, y_train)
validation_pool = catboost.Pool(X_val, y_val)

model = CatBoostClassifier(
                            custom_loss=['AUC'],
                            random_seed=42,
                            logging_level='Silent'
                            )
model.fit(
    train_pool,
    eval_set=validation_pool,
    plot=True
)
```

```python
plt.bar(df.columns, model.feature_importances_)
plt.xticks(rotation=90)
plt.show()
```

```python
print('Training set metrics:')
print('Accuracy:', accuracy_score(y_train, model.predict(X_train)))
print('Precision:', precision_score(y_train, model.predict(X_train)))
print('Recall:', recall_score(y_train, model.predict(X_train)))
print('-------------------------')
print('Test set metrics:')
print('Accuracy:', accuracy_score(y_test, model.predict(X_test)))
print('Precision:', precision_score(y_test, model.predict(X_test)))
print('Recall:', recall_score(y_test, model.predict(X_test)))

plot_confusion_matrix(model,
                     X_test,
                     y_test,
                     values_format='d',
                      display_labels=['non embauché','embauché'])
```

Maintenant nous allons appliquer le modèle XGboost directement. En suite, nous allons optimisé ses paramètres avec une démarche expliquée dans ce qui suit.

#### Application du modèle XGBoost

```python
# load data
model = XGBClassifier(Objective='binary:logistic',seed=42)
model.fit(X_train,
          y_train,
          early_stopping_rounds=10,
          verbose=0,
          eval_metric='aucpr',
          eval_set=[(X_test,y_test)]
         )
y_test_hat = model.predict(X_test)

print('Test performance')
print('-------------------------------------------------------')
print(classification_report(y_test, y_test_hat))
plot_confusion_matrix(model,
                     X_test,
                     y_test,
                     values_format='d',
                      display_labels=['non embauché','embauché'])
```

#### Optimisation des hyperparamètres

Quelques recommandation trouvés dans le manuel de XGBoost: 

" Traitement d'un ensemble de données déséquilibré

Pour des cas courants tels que le journal des clics publicitaires, l'ensemble de données est extrêmement déséquilibré. Cela peut affecter la formation du modèle XGBoost, et il y a deux façons de l'améliorer.
Si vous ne vous souciez que de la métrique de performance globale (AUC) de votre prédiction

        Équilibrez les poids positifs et négatifs via scale_pos_weight.
        Utilisez l'AUC pour l'évaluation "
Ce que nous allons faire :
 - Définir des paramètres à optimiser 
 - Faire un cross validation de ces paramètres
 - Effectuer un mise-à-jour de ces paramètres selon les résultats trouvés
 - Pour une liste de paramètre donnée, si la valeur du paramètre optimisé (après grid search) se trouve au bord on élargie la liste, sinon on fixe le paramètre pour le deuxième grid search

```python
# First round params

param_grid={
     "gamma"               : [ 0.0, 0.25, 1.0  ],  # Paramètre qui encourage le prunning
    
     "learning_rate"       : [ 0.0, 0.10, 0.15 ] , # Le pas de l'optimisateur
    
     "max_depth"           : [ 3, 4, 5],           # Profondeur maximale d'un arbre. 
    
     "reg_lambda"          : [0, 1.0, 10.0],       # Paramètre de régularisation
    
     "scale_pos_weight"    : [ 1, 3 ]              # Tenir en compte de la déséquilibration de data
    }


model = XGBClassifier(Objective='binary:logistic', # Problème de classification binaire
                      seed=42,
                      subsample=0.9,               # A chaque construction d'arbre un sous-ensemble aléatoire de 90% 
                                                   # sera choisi
                      
                      colsample_bytree=0.5)        # Pour chaque arbre un choix aléatoire de 50% des colonnes sera réalisé
                                                   # Ces deux paramètres permettent d'éviter l'overfitting 
                                                   # et d'accélèrer la validation croisée
                     
grid_search = GridSearchCV(model,
                           param_grid=param_grid,
                           scoring="roc_auc",      # Recommander par le manuel
                           verbose=0,
                           n_jobs=-1,              # Utiliser tous les noyaux existants
                           cv=3)
grid_result = grid_search.fit(X_train, y_train)
```

```python
grid_result.best_params_
```

* Nous allons maintenant modifier notre liste de valeurs de paramètres en fonction des résultats de la validation croisée. Pour chaque paramètre, si la valeur résultante est au milieu de la liste de ses choix de valeurs, nous la fixons à cette valeur, sinon nous poussons ses choix de valeurs si la valeur résultante est au bord de la liste.

```python
# First round
param_grid={
    "gamma"               : [ 1.0, 3.0, 5.0  ],  # pousser la liste pour des valuers plus grandes que la précédante
    
    "learning_rate"       : [ 0.15, 0.2, 0.25 ], 
    
    "max_depth"           : [ 5, 7, 9],    
                                            
    "reg_lambda"          : [ 10.0, 13, 15], 
     
    "scale_pos_weight"    : [ 0.1, 0.5, 1 ]      # pousser la liste pour des valuers plus faibles que la précédante
    }

grid_search = GridSearchCV(model,
                           param_grid=param_grid,
                           scoring="roc_auc",
                           verbose=0,
                           n_jobs=-1,
                           cv=3)
grid_result = grid_search.fit(X_train, y_train)
grid_result.best_params_
```

```python
# Second round

param_grid={
         "gamma"               : [5.0, 7.0, 9.0], 

         "learning_rate"       : [0.25, 0.30, 0.35 ] , 
    
         "max_depth"           : [6, 7, 8],    
    
         "reg_lambda"          : [12.0, 13, 14.0], 
    
         "scale_pos_weight"    : [ 1 ]  }  # Fixer cette valeur 

grid_search = GridSearchCV(model,
                           param_grid=param_grid,
                           scoring="roc_auc",
                           verbose=0,
                           n_jobs=-1,
                           cv=3)
grid_result = grid_search.fit(X_train, y_train)
grid_result.best_params_
```

```python
param_grid={
         "gamma"               : [5.0],  # Fixer cette valeur 

         "learning_rate"       : [0.25] , 
    
         "max_depth"           : [7],    
                                        
         "reg_lambda"          : [13], 
    
         "scale_pos_weight"    : [1]   
    }

model = XGBClassifier(Objective='binary:logistic',  
                      seed            = 42,
                      subsample       = 0.9,
                      colsample_bytree= 0.7,
                      gamma           = 5.0,
                      learning_rate   = 0.25,
                      max_depth       = 7,
                      reg_lambda      = 13.0,
                      scale_pos_weight= 1
                     )
model.fit(X_train,
          y_train,
          early_stopping_rounds=30,
          verbose=0,
          eval_metric='aucpr',
          eval_set=[(X_test,y_test)]
         )

plot_confusion_matrix(model,
                     X_test,
                     y_test,
                     values_format='d',
                      display_labels=['non embauché','embauché'])
y_test_hat = model.predict(X_test)

print('Test performance')
print('-------------------------------------------------------')
print(classification_report(y_test, y_test_hat))
```

```python
print('Training set metrics:')
print('Accuracy:', accuracy_score(y_train, model.predict(X_train)))
print('Precision:', precision_score(y_train, model.predict(X_train)))
print('Recall:', recall_score(y_train, model.predict(X_train)))
print('-------------------------')
print('Test set metrics:')
print('Accuracy:', accuracy_score(y_test, model.predict(X_test)))
print('Precision:', precision_score(y_test, model.predict(X_test)))
print('Recall:', recall_score(y_test, model.predict(X_test)))
```
